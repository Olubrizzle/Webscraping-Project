```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import re
```

```python
data1 = pd.read_csv('Avandia.csv')
data2 = pd.read_csv('Metformin.csv')
data3 = pd.read_csv('Invokana.csv')
```

```python
list(data1.columns)
```

```python
data1['drug'] = 'Avandia'
data1.head()
```

```python
data2['drug'] = 'Metformin'
data2.head()
```

```python
data3['drug'] = 'Invokana'
data3.head()
```

```python
merged_table = pd.concat([data1, data2, data3])
merged_table.head()
```

```python
def dur(text):
    new_duration = ''.join(re.findall('[0-9]+',text))
    if new_duration == '1':
        return .1
    elif new_duration == '16':
        return .3
    elif new_duration == '61':
        return 0.75
    elif new_duration == '12':
        return 1.5
    elif new_duration == '25':
        return 3.5
    elif new_duration == '510':
        return 7.5
    else:
        return ''
merged_table['adjusted_duration'] = [dur(x) for x in merged_table['duration']]
merged_table.sample()
```

```python
group1 = merged_table.groupby(['drug','effectiveness_ratings']).agg('count')
group1
```
